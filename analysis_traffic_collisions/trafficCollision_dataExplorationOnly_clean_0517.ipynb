#### Columns
OBJECTID - ESRI unique identifier

ADDRTYPE - Text, 12 VARCHAR, Collision address type:

$\,\,$• Block  
$\,\,$• Intersection

SEVERITYCODE - code that corresponds to the severity of the collision:  
$\,\,$• 3—fatality  
$\,\,$• 2b—serious injury  
$\,\,$• 2—injury  
$\,\,$• 1—prop damage  
$\,\,$• 0—unknown  
INCDATE - date of the incident  
INCDTTM - date and time of the incident, Text, 30 VARCHAR  
JUNCTIONTYPE - Category of junction at which collision took place  
SDOT_COLCODE - A code given to the collision by SDOT  
SDOT_COLDESC - A description of the collision corresponding to the collision code  
INATTENTIONIND - Whether or not collision was due to inattention. (Y/N)  
PEDROWNOTGRNT - Whether or not the pedestrian right of way was not granted. (Y/N)

$\,\,$0 Vehicle Going Straight Hits Pedestrian

$\,\,$10 Entering At Angle   
$\,\,$11 From Same Direction -Both Going Straight-Both Moving- Sideswipe  
$\,\,$12 From Same Direction -Both Going Straight-One Stopped- Sideswipe  
$\,\,$13 From Same Direction - Both Going Straight - Both Moving - Rear End  
$\,\,$...

```python
now = datetime.datetime.now()
url = "https://data-seattlecitygis.opendata.arcgis.com/datasets/5b5c745e0f1f48e7a53acec63a0022ab_0.csv"
df = pd.read_csv(url)
print("df shape: {}/n".format(df.shape))
print(df.describe())

df.to_csv("collision_data_{}{}{}.csv".format(now.month, now.day, now.year), encoding="utf-8", index=False)

df.head()
```

```python
# missing data

df_na = df.isna().sum().to_frame("total_null")
df_na["percent_null"] = np.round((df_na["total_null"]/len(df))*100, 0)
df_na["total_unique_values"] = df.nunique()
display(df_na)
display(df_na.sort_values("percent_null", ascending=True).head(10))
```

```python
# relabel boolean columns
df["speeding_bool"] = df["SPEEDING"].map({"Y": 1, np.nan: 0}).astype("category") #### ?????
display(df["SPEEDING"].value_counts())
display(df["speeding_bool"].value_counts())

# break out date for later analysis, but keep the datetime column
df["month"] = pd.DatetimeIndex(df['INCDATE']).month
df["year"] = pd.DatetimeIndex(df['INCDATE']).year
df["day_of_week"] = df['INCDATE'].apply(lambda x: pd.Timestamp(x).dayofweek)
df_month = df.groupby(['month']).size().to_frame('count')
df_month["percent"] = np.round((df_month["count"]/len(df))*100, 0)

df['SEVERITYSTR'] = df['SEVERITYCODE'].map({'0': 'unknown', '1': 'prop_damage',
                                            '2': 'injury', '2b': 'serious_injury', 
                                            '3': 'fatality'})
display(df.groupby(['SEVERITYSTR']).size().to_frame('count'))
```

```python
# there are a lot of missing location integers in "INTKEY"
print("Total missing values in df['INTKEY'] = {}%\n".format(np.round(((df["INTKEY_4"].value_counts()[0]/len(df))*100), 0)))

# of the labeled locations, it's spread out, so there's not a clear pattern
print(df.groupby('INTKEY').size().to_frame("count").sort_values("count", ascending=False).head(20), "\n") #[0:25]) sort_values(axis=1)

# let's zoom out to larger areas to see if the pattern changes
df["INTKEY_4"] = df["INTKEY"].fillna(value=0, axis=0)
df["INTKEY_4"] = df["INTKEY_4"].astype(str).str[:4]
df["INTKEY_4"] = pd.to_numeric(df["INTKEY_4"], downcast="integer")
print(df["INTKEY_4"].value_counts().to_frame("count").head(20))
```
